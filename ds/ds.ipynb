```python
import numpy as np

from bokeh.plotting import figure, output_notebook, show
  
import datashader as ds
import datashader.transfer_functions as tf
from datashader.callbacks import InteractiveImage

from functools import partial
  
import pandas as pd
output_notebook()
```

```python
df = pd.read_csv("../test/sample.csv", usecols=['ra', 'dec', 'mag_auto_i'])
  
x_range=(np.min(df['ra']),np.max(df['ra']))
y_range=(np.min(df['dec']),np.max(df['dec']))
  
```

```python
def base_plot():
    p = figure(tools='pan,wheel_zoom,box_zoom,reset', 
                 plot_width=2000, plot_height=300, 
                 x_range=x_range, y_range=y_range)
    p.axis.visible = True
    return p
  
p = base_plot()
```

```python
def create_image(x_range, y_range, w, h, color_fn=partial(tf.interpolate,low="lightblue",high="darkred",how='log')):
      cvs = ds.Canvas(plot_width=w, plot_height=h, x_range=x_range, y_range=y_range)
      agg = cvs.points(df, 'ra',  'dec',  ds.count('dec'))
      image = color_fn(agg)
      return image
  
InteractiveImage(p, create_image)
```
